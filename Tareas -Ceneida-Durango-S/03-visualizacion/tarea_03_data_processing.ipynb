## Parte 1 — Exploración y selección

**P1.** Imprime los nombres de todas las columnas y el número total de filas y columnas del DataFrame.

```python
from IPython.display import display

tabla_columnas = pd.DataFrame({"Columnas": df.columns})
tabla_dimensiones = pd.DataFrame({"Filas": [df.shape[0]], "Columnas": [df.shape[1]]})

display(tabla_columnas)
display(tabla_dimensiones)
```

## Parte 2 — Valores nulos y limpieza

**P3.** Cuenta cuántos valores nulos hay en cada columna. ¿Cuáles columnas tienen nulos?

```python
nulos_por_columna = df.isnull().sum()
columnas_con_nulos = nulos_por_columna[nulos_por_columna > 0]

print(columnas_con_nulos.index.tolist())
columnas_con_nulos
```

**Respuesta:** Las columnas con valores nulos son: Age, Cabin y Embarked.

**P4.** Rellena los valores nulos de la columna `Age` con la **mediana** de esa columna.  
Verifica que ya no haya nulos en `Age`.

```python
mediana_age = df["Age"].median()
df["Age"] = df["Age"].fillna(mediana_age)

print("Nulos en Age después del relleno:", df["Age"].isnull().sum())
```

---
## Parte 3 — Nuevas columnas y agrupaciones

**P5.** Crea una nueva columna llamada `FamilySize` que sea la suma de `SibSp` + `Parch` + 1 (el pasajero mismo).

```python
df["FamilySize"] = df["SibSp"] + df["Parch"] + 1
df.head()
```

**P6.** Usando `groupby`, calcula la **tasa de supervivencia promedio** (columna `Survived`) por clase (`Pclass`).  
¿Qué clase tuvo mayor probabilidad de sobrevivir?

```python
supervivencia_por_clase = df.groupby("Pclass")["Survived"].mean()

print(supervivencia_por_clase)
print("La clase con mayor supervivencia fue:", supervivencia_por_clase.idxmax())
```

**P7.** Calcula la edad promedio y la tarifa (`Fare`) promedio, agrupando por `Sex` y `Pclass`.  
Usa `.groupby()` con `.agg()`.

```python
resumen = (
    df.groupby(["Sex", "Pclass"])
    .agg({"Age": "mean", "Fare": "mean"})
    .rename(columns={"Age": "Edad_promedio", "Fare": "Tarifa_promedio"})
)

resumen
```

## Parte 4 — Ordenamiento y duplicados

**P8.** Ordena el DataFrame por `Fare` de mayor a menor y muestra las primeras 5 filas.  
¿Cuánto pagó el pasajero más caro?

```python
df_ordenado = df.sort_values(by="Fare", ascending=False)

df_ordenado.head()

print("El pasajero más caro pagó:", df_ordenado["Fare"].iloc[0])
```

**P9.** Verifica si hay filas duplicadas en el DataFrame. ¿Cuántas hay?

```python
duplicadas = df.duplicated().sum()
print("Cantidad de filas duplicadas:", duplicadas)
print("¿Hay filas duplicadas?", df.duplicated().any())
print("Cantidad de filas duplicadas:", df.duplicated().sum())
```
