## 3. Cargar tablas a SQLite

### Punto 1. Selección y filtro básico

Trae todos los estudiantes que viven en **Bogotá**.

### Punto 3. Filtro con múltiples condiciones

### Punto 4. Agregación simple

Calcula:
- número total de cursos
- precio promedio
- precio mínimo
- precio máximo

Usa la tabla `cursos`.

**Entregable:**
- SQL
- pandas

```python
# Punto 4 - SQL
query_4 = """
SELECT COUNT(*) AS total_cursos, AVG(precio) AS precio_promedio, MIN(precio) AS precio_minimo, MAX(precio) AS precio_maximo
FROM cursos
"""
sql_4 = run_sql(query_4)
sql_4
```

```python
# Punto 4 - pandas
cursos["precio"].agg(
    total_cursos="count",
    precio_promedio="mean",
    precio_minimo="min",
    precio_maximo="max",
)
```

### Punto 5. GROUP BY

Cuenta cuántos cursos hay por `categoria`.

Ordena de mayor a menor cantidad.

**Entregable:**
- SQL
- pandas

```python
# Punto 5 - SQL
query_5 = """
SELECT categoria, COUNT(*) AS cantidad_cursos
FROM cursos
GROUP BY categoria
ORDER BY cantidad_cursos DESC
"""
sql_5 = run_sql(query_5)
sql_5
```

### Punto 6. JOIN entre tablas

Muestra:
- nombre del estudiante
- nombre del curso
- estado de la inscripción

Usa `estudiantes`, `inscripciones` y `cursos`.

**Entregable:**
- SQL
- pandas

```python
# Punto 6 - SQL
query_6 = """
SELECT e.nombre AS nombre_estudiante, c.nombre_curso, i.estado
FROM estudiantes e
JOIN inscripciones i ON e.estudiante_id = i.estudiante_id
JOIN cursos c ON i.curso_id = c.curso_id
"""
sql_6 = run_sql(query_6)
sql_6
```

```python
# Punto 6 - pandas
inscripciones.merge(estudiantes, on="estudiante_id").merge(cursos, on="curso_id")[
    ["nombre", "nombre_curso", "estado"]
].rename(columns={"nombre": "nombre_estudiante"})
```

### Punto 7. LEFT JOIN

Trae todos los estudiantes, incluso si no tuvieran inscripciones, mostrando:
- `estudiante_id`
- `nombre`
- `inscripcion_id`

**Entregable:**
- SQL
- pandas

> Pista: en este dataset puede que todos tengan al menos una inscripción, pero igual debes practicar el `LEFT JOIN`.

```python
# Punto 7 - SQL
query_7 = """
SELECT e.estudiante_id, e.nombre, i.inscripcion_id
FROM estudiantes e
LEFT JOIN inscripciones i
on e.estudiante_id = i.estudiante_id
"""
sql_7 = run_sql(query_7)
sql_7
```

```python
# Punto 7 - pandas
estudiantes.merge(inscripciones, on="estudiante_id", how="left")[
    ["estudiante_id", "nombre", "inscripcion_id"]
]
```

### Punto 8. HAVING

Calcula cuántas inscripciones tiene cada estudiante y muestra solo aquellos con **2 o más**.

Debes mostrar:
- `estudiante_id`
- `nombre`
- `n_inscripciones`

**Entregable:**
- SQL
- pandas

```python
# Punto 8 - SQL
query_8 = """
SELECT
 e.estudiante_id,
 e.nombre,
 COUNT(i.inscripcion_id) AS n_inscripciones
FROM estudiantes e
JOIN inscripciones i
 ON e.estudiante_id = i.estudiante_id
GROUP BY e.estudiante_id, e.nombre
HAVING COUNT (i.inscripcion_id) >= 2
"""
sql_8 = run_sql(query_8)
sql_8
```

```python
# Punto 8 - pandas
inscripciones.groupby("estudiante_id").size().reset_index(name="n_inscripciones").merge(
    estudiantes, on="estudiante_id"
).query("n_inscripciones >= 2")[["estudiante_id", "nombre", "n_inscripciones"]]
```

### Punto 9. CASE WHEN / segmentación

Clasifica los cursos según su precio:
- `Bajo` si el precio es menor a 120
- `Medio` si el precio está entre 120 y 180 inclusive
- `Alto` si el precio es mayor a 180

Debes mostrar:
- `nombre_curso`
- `precio`
- `segmento_precio`

**Entregable:**
- SQL
- pandas

```python
query_9 = """
SELECT
    nombre_curso,
    precio,
    CASE
        WHEN precio < 120 THEN 'Bajo'
        WHEN precio BETWEEN 120 AND 180 THEN 'Medio'
        ELSE 'Alto'
    END AS segmento_precio
FROM cursos
"""
sql_9 = run_sql(query_9)
sql_9
```

```python
cursos.assign(
    segmento_precio=cursos["precio"].apply(
        lambda x: "Bajo" if x < 120 else "Medio" if x <= 180 else "Alto"
    )
)[["nombre_curso", "precio", "segmento_precio"]]
```

### Punto 10. Ejercicio final integrador

Calcula el total pagado por categoría de curso.

Debes:
- unir `cursos`, `inscripciones` y `pagos`
- agrupar por `categoria`
- sumar `monto`
- ordenar de mayor a menor

**Entregable:**
- SQL
- pandas

```python
query_10 = """
SELECT
    c.categoria,
    SUM(p.monto) AS total_pagado
FROM cursos c
JOIN inscripciones i
    ON c.curso_id = i.curso_id
JOIN pagos p
    ON i.inscripcion_id = p.inscripcion_id
GROUP BY c.categoria
ORDER BY total_pagado DESC
"""
sql_10 = run_sql(query_10)
sql_10
```

```python
total_pagado_categoria = (
    inscripciones.merge(cursos, on="curso_id")
    .merge(pagos, on="inscripcion_id")
    .groupby("categoria", as_index=False)["monto"]
    .sum()
    .rename(columns={"monto": "total_pagado"})
    .sort_values(by="total_pagado", ascending=False)
)

total_pagado_categoria
```
